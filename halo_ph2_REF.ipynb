```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet, SGDRegressor
from sklearn.cross_validation import KFold
from sklearn.preprocessing import StandardScaler,scale
from sklearn import metrics

# 'EV1_FLAG','EV2_FLAG','EV3_FLAG','EV4_FLAG','EV5_FLAG',,'FIXP_FLAG','FIXP_FLAG'

from sklearn.preprocessing import Imputer
imp = Imputer(missing_values='NaN', strategy='mean', axis=0)
impt = Imputer(missing_values='NaN', strategy='mean', axis=0)

Total = pd.DataFrame()

for stores in df.STORE_ID.unique():
    print ('running for' +str(stores))
    X = df[df['STORE_ID'] == stores]
    X['FEAT_LOG'] = np.log(X.FEAT_FLAG )
    X['FIXP_LOG'] = np.log(X.FIXP_FLAG )
    X['NONP_LOG'] = np.log(X.NONP_FLAG )
    X['FEAT_PART_LOG'] = np.log(X.FEAT_PART_FLAG )
    X['FIXP_PART_LOG'] = np.log(X.FIXP_PART_FLAG )
    X['NONP_PART_LOG'] = np.log(X.NONP_PART_FLAG )

    sub = ['STORE_ID','DATE_ID','TOTAL_UNITS','TOTAL_SALES','REST_ST_UNITS','NONP_FLAG','NONP_PART_FLAG','FEAT_PART_FLAG','FIXP_PART_FLAG',
          'EV1_FLAG','EV2_FLAG','EV3_FLAG','EV4_FLAG','EV5_FLAG','EV6_FLAG','EV7_FLAG','EV8_FLAG','EV9_FLAG','EV10_FLAG',
           'WKND_FLAG','FEAT_PART_LOG','FIXP_PART_LOG',
          'FEAT_FLAG','FIXP_FLAG','NONP_PART_LOG','NONP_SALES','NONP_UNITS']
# 'FEAT_PART_LOG','FIXP_PART_LOG','HIGH_TEMP','LOW_TEMP','WIND_BFS','MAX_CLOUD_COVER','DAY_YR',
    Y_train = (X[X.DAY_FLAG < 500]).REST_ST_SALES
    Y_test = (X[X.DAY_FLAG >= 500]).REST_ST_SALES

    X.drop(sub, axis=1, inplace=True)
    X.drop('REST_ST_SALES', axis=1, inplace=True)
    X_tr = X[X.DAY_FLAG < 500]
    X_tr.drop('DAY_FLAG', axis = 1, inplace = True)
    X_tst = X[X.DAY_FLAG >= 500]   
    X_tst.drop('DAY_FLAG', axis = 1, inplace = True)

    X_scaler = StandardScaler()
    X_tr = X_tr.replace([np.inf, -np.inf], np.nan)
    X_tr = X_tr.fillna(0)
    X_tst = X_tst.replace([np.inf, -np.inf], np.nan)
    X_tst = X_tst.fillna(0)
    X_train = X_scaler.fit_transform(X_tr)
    X_test = X_scaler.transform(X_tst)

    Y_train = Y_train.fillna(0)
    Y_test = Y_test.fillna(0)

    
#     X_train = X_scaler.fit_transform(imp.fit_transform(X_tr.as_matrix()))
#     X_test = X_scaler.transform(impt.fit_transform(X_tst.as_matrix()))


    alpha = np.logspace(-2,6,250)
    t_rmse = np.array([])
    cv_rmse = np.array([])

    for a in alpha:
        rg_model = ElasticNet(alpha = a,normalize=False,l1_ratio=0)
        rg_model.fit(X_train,Y_train)
        train_pred= rg_model.predict(X_train)
        Predicted_price= rg_model.predict(X_test)
    # calculate MAE using scikit-learn
        rmse_train = np.sqrt(metrics.mean_squared_error(Y_train, train_pred))
        rmse_test = np.sqrt(metrics.mean_squared_error(Y_test, Predicted_price))
        t_rmse = np.append(t_rmse, [rmse_train])
        cv_rmse = np.append(cv_rmse, [rmse_test])
#         print('{:.3f}\t {:.4f}\t\t {:.4f}'.format(a,rmse_train,rmse_test))

    str_alpha = pd.DataFrame()
    str_alpha = pd.concat([pd.DataFrame(alpha),pd.DataFrame(cv_rmse)], axis = 1) 
    new_alpha = np.mean(str_alpha[str_alpha.iloc[:,1] == cv_rmse.min()].iloc[:,0])

    rg_model = ElasticNet(alpha = new_alpha,normalize=False,l1_ratio=0)
    rg_model.fit(X_train,Y_train)
    print('Ridge Regression')
    train_pred= rg_model.predict(X_train)
    Predicted_price= rg_model.predict(X_test)
    # calculate MAE using scikit-learn
    print("Rsquared value for training: " + str(round(rg_model.score(X_train,Y_train),3)))

    print("Rsquared value for test: " + str(round(rg_model.score(X_test,Y_test),3)))
    
    print("RMSE value for training: " + str(np.sqrt(metrics.mean_squared_error(Y_train, train_pred))))
    print("RMSE value for test: " + str(np.sqrt(metrics.mean_squared_error(Y_test, Predicted_price))))
    
    print("alpha value: " + str(new_alpha))

    Coeff = pd.DataFrame(X_tr.columns, rg_model.coef_)
    print(Coeff)
    print(rg_model.intercept_)
    
    D = pd.DataFrame(pd.concat([pd.DataFrame(X_train),pd.DataFrame(X_test)])*rg_model.coef_)
    D.columns = X_tr.columns
    
    tr_res = pd.DataFrame({"store_id":stores, "Observed":Y_train,"Predicted":train_pred,"intercept":rg_model.intercept_,
                           "day":X[X.DAY_FLAG < 500].DAY_FLAG})
    test_res = pd.DataFrame({"store_id":stores, "Observed":Y_test,"Predicted":Predicted_price,"intercept":rg_model.intercept_,
                             "day":X[X.DAY_FLAG >= 500].DAY_FLAG})
    Result = pd.DataFrame({"Observed":Y_test,"Predicted":Predicted_price})
# ,"day" :X_tr.DAY_FLAG
    Result = pd.concat([pd.concat([tr_res,test_res]).reset_index(),D.reset_index()],axis =1)
    Total = pd.concat([Total,Result])

Total.to_csv('/nfs/science/shared/ipythonNotebooks/anantk/halo_output_08_02_all.csv')  


```
