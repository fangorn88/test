```python
bdata = pd.read_csv('/nfs/science/shared/ipythonNotebooks/garimaj/McDonalds/nmf_base_data.csv',header=0, delimiter=",")
df=bdata.groupby(['trans_key','item_name'])['units'].max().unstack()

print df.shape
```

```python
import numpy as np

X = df.as_matrix()
print X.view()

where_are_NaNs = np.isnan(X)
X[where_are_NaNs] = 0
print X.view()
print X.shape
```

## Run 1

```python
from sklearn.decomposition import NMF
import time

start_time=time.time()

c=5  
    
print("Fitting the NMF model for %d features" %(c) )
nmf = NMF(n_components=c, beta_loss='kullback-leibler',random_state=12, alpha=0.1,verbose =2, max_iter=300,
          l1_ratio=0.5,init = 'nndsvda', solver='mu')
nmf.fit(X)

time_taken=time.time()-start_time
minutes, seconds=divmod(time_taken,60)
print("Time taken to complete: %d minutes, %d seconds" %(minutes,seconds))

print "Error: %d" %(nmf.reconstruction_err_)
print "Iterations: %d" %(nmf.n_iter_)

start_time=time.time()

c=6
    
print("Fitting the NMF model for %d features" %(c) )
nmf6 = NMF(n_components=c, beta_loss='kullback-leibler',random_state=12, alpha=0.1,verbose =2, max_iter=300,
          l1_ratio=0.5,init = 'nndsvda', solver='mu')
nmf6.fit(X)

time_taken=time.time()-start_time
minutes, seconds=divmod(time_taken,60)
print("Time taken to complete: %d minutes, %d seconds" %(minutes,seconds))

print "Error: %d" %(nmf6.reconstruction_err_)
print "Iterations: %d" %(nmf6.n_iter_)

c=7
    
print("Fitting the NMF model for %d features" %(c) )
nmf7 = NMF(n_components=c, beta_loss='kullback-leibler',random_state=12, alpha=0.1,verbose =2, max_iter=300,
          l1_ratio=0.5,init = 'nndsvda', solver='mu')
nmf7.fit(X)

time_taken=time.time()-start_time
minutes, seconds=divmod(time_taken,60)
print("Time taken to complete: %d minutes, %d seconds" %(minutes,seconds))

print "Error: %d" %(nmf7.reconstruction_err_)
print "Iterations: %d" %(nmf7.n_iter_)

c=8
    
print("Fitting the NMF model for %d features" %(c) )
nmf8 = NMF(n_components=c, beta_loss='kullback-leibler',random_state=12, alpha=0.1,verbose =2, max_iter=300,
          l1_ratio=0.5,init = 'nndsvda', solver='mu')
nmf8.fit(X)

time_taken=time.time()-start_time
minutes, seconds=divmod(time_taken,60)
print("Time taken to complete: %d minutes, %d seconds" %(minutes,seconds))

print "Error: %d" %(nmf8.reconstruction_err_)
print "Iterations: %d" %(nmf8.n_iter_)
```

```python
# Baskets X Missions matrix
bask_tag_12=nmf12.transform(X) #(n_samples, n_components)
bask_tag_12.shape
```

```python
bask_tg_12=pd.DataFrame(bask_tag_12)
bask_tg_12.index=df.index
bask_tg_12.head()
```

```python
# pd.set_option('display.float_format', lambda x: '%.5f' % x)
del bask_tg_12['max']
del bask_tg_12['max_2']

tagging=pd.DataFrame()
tagging['max']=bask_tg_12.apply(lambda x: bask_tg_12.columns[x.argsort()[::-1][11]], axis=1)
tagging['max_2']=bask_tg_12.apply(lambda x: bask_tg_12.columns[x.argsort()[::-1][10]], axis=1)
tagging.head(20)
```

```python
bask_tg_12=bask_tg_12.merge(tagging, how='left', left_index=True, right_index=True, sort=False, suffixes=('_x', '_y'), copy=False)
bask_tg_12.head()
```
