```python
import numpy as np
import pandas as pd
import csv as csv
import zipfile

home_train = pd.read_csv('/nfs/science/shared/ipythonNotebooks/anantk/Kgl/Home/train.csv',  header=0, sep=',')
print(home_train.head())

home_test = pd.read_csv('/nfs/science/shared/ipythonNotebooks/anantk/Kgl/Home/test.csv',  header=0, sep=',')
# print(home_test.head())

# home_train

all_data = pd.concat((home_train.loc[:,'MSSubClass':'SaleCondition'],
                      home_test.loc[:,'MSSubClass':'SaleCondition']))
```

```python
from sklearn.preprocessing import StandardScaler, scale

all_data = pd.get_dummies(all_data)

print(all_data.head())

all_data = all_data.fillna(0)

scaler = StandardScaler()
alldata = pd.DataFrame(scaler.fit_transform(all_data))
alldata.columns = all_data.columns
# X_test = scaler.transform(X_test)

X_train = pd.DataFrame(alldata[:home_train.shape[0]])
X_test = pd.DataFrame(alldata[home_train.shape[0]:])

Y_train = home_train.SalePrice

X_train
```

```python
import time
from operator import itemgetter
from scipy.stats import randint as sp_randint

from sklearn.datasets import load_digits
from sklearn.grid_search import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

# rf = RandomForestRegressor(random_state=30)
# rf.fit(X_train,Y_train)

# print("mse_train: " + str(metrics.mean_squared_error(Y_train,rf.predict(X_train))))
# print("mse_test: " + str(metrics.mean_squared_error(Y_test,rf.predict(X_test))))

# Utility function to report best scores

# def report(grid_scores, n_top=3):
#     top_scores = sorted(grid_scores, key=itemgetter(1), reverse=True)[:n_top]
#     for i, score in enumerate(top_scores):
#         print("Model with rank: {0}".format(i + 1))
#         print("Mean validation score: {0:.3f} (std: {1:.3f})".format(
#               score.mean_validation_score,
#               np.std(score.cv_validation_scores)))
#         print("Parameters: {0}".format(score.parameters))
#         print("")

# model = RandomForestRegressor(random_state=30)
# param_grid = { "n_estimators"      : [50,250,500],
#            "max_features"      : ['auto','sqrt','log2'],
#            "max_depth"         : [4,10, 20],
#            "min_samples_split" : [1, 2, 4] ,
#            "bootstrap": [True, False]}
# grid_search = GridSearchCV(model, param_grid, cv=3, verbose = 15)
# start = time.time()
# grid_search.fit(X_train, Y_train)

# print("GridSearchCV took %.2f seconds for %d candidate parameter settings."
#       % (time.time() - start, len(grid_search.grid_scores_)))
# report(grid_search.grid_scores_)
# print(grid_search.best_params_)


def report(grid_scores, n_top=10):
    top_scores = sorted(grid_scores, key=itemgetter(1), reverse=True)[:n_top]
    for i, score in enumerate(top_scores):
        print("Model with rank: {0}".format(i + 1))
        print("Mean validation score: {0:.3f} (std: {1:.3f})".format(
              score.mean_validation_score,
              np.std(score.cv_validation_scores)))
        print("Parameters: {0}".format(score.parameters))
        print("")

model = RandomForestRegressor(random_state=30)
param_grid = { "n_estimators"      : [250,500,750,1000,1250],
           "max_features"      : ['sqrt'],
           "max_depth"         : [20,30,40],
           "min_samples_split" : [1, 2, 4] ,
           "bootstrap": [True, False]}
grid_search = GridSearchCV(model, param_grid, cv=5, verbose = 5)
start = time.time()
grid_search.fit(X_train, Y_train)

print("GridSearchCV took %.2f seconds for %d candidate parameter settings."
      % (time.time() - start, len(grid_search.grid_scores_)))
report(grid_search.grid_scores_)
print(grid_search.best_params_)
```
