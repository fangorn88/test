```python
# Importing the libraries for basic data operations

import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt

%matplotlib inline

# Loading the famous Boston housing data set to be used in the training

from sklearn.datasets import load_boston

Boston = load_boston()

print(Boston.DESCR)
print(Boston.target[0:40,])

boston = pd.DataFrame(Boston.data)

boston.columns = Boston.feature_names
boston.head()
```

```python
import seaborn as sns
sns.set()

sns.pairplot(boston)

def draw_histograms(df, variables, n_rows, n_cols):
    fig=plt.figure()
    for i, var_name in enumerate(variables):
        ax=fig.add_subplot(n_rows,n_cols,i+1)
        df[var_name].hist(bins=10,ax=ax)
        ax.set_title(var_name+" Distribution")
    fig.tight_layout()  # Improves appearance a bit.
    plt.show()

draw_histograms(boston, boston.columns, 4, 4)

def draw_scatterplot(df,variables,n_rows,n_cols):
    fig=plt.figure()
    for i,var_name in enumerate(variables):
        plt.subplot(n_rows,n_cols,i+1)
        plt.scatter(df.iloc[:,i], Boston.target)
        plt.xlabel(var_name)
        plt.ylabel('Target')
        plt.title(var_name+' with target')
    fig.tight_layout()  # Improves appearance a bit.   
    plt.show()

draw_scatterplot(boston, boston.columns, 4, 4)
```
